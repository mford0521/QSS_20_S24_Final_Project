## This file contains code for heatmap visualizations 

```python
# import necesary packages 
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load demographic abc data 
demographic_abcs = pd.read_csv('../data/demographic_abcs.csv')
```

```python
# functions to create heat maps by demographic type, organized left to right by mean ABC Score change

# melt the dataframe 
def reshape_data(abc_grouped, id_vars):
    return abc_grouped.melt(
        id_vars=id_vars, 
        value_vars=['norm_leth_soc', 'norm_ir_agit', 'norm_stereo', 'norm_hyper', 'norm_speech'],
        var_name='ABC Score', value_name='Mean Value'
    )

# calculate changes in ABC scores from intake to re-eval
def calculate_differences(abc_pivot):
    return abc_pivot.loc['Re-evaluation'] - abc_pivot.loc['Intake']

# sort by difference/improvement in ABC 
def rename_and_sort_index(abc_diff, rename_dict):
    new_index = [rename_dict[abc_score] for abc_score in abc_diff.index]
    abc_diff.index = new_index
    mean_abc_scores_diff = abc_diff.mean().sort_values(ascending=True)
    return abc_diff[mean_abc_scores_diff.index]

# plt heatmap 
def plot_heatmap(abc_diff, title, x_label, y_label, output_file):
    plt.figure(figsize=(18, 10))
    heatmap = sns.heatmap(
        abc_diff,
        annot=False,
        cmap='viridis',
        cbar_kws={'shrink': 0.5}
    )
    heatmap.set_xticklabels(heatmap.get_xticklabels(), fontsize=14, fontweight='bold')
    heatmap.set_yticklabels(heatmap.get_yticklabels(), fontsize=14, fontweight='bold')
    heatmap.tick_params(left=False, bottom=False)
    plt.title(title, fontsize=16, fontweight='bold')
    plt.xlabel(x_label, fontsize=14, fontweight='bold')
    plt.ylabel(y_label, fontsize=14, fontweight='bold')
    plt.savefig(output_file)
    plt.show()
```

```python
# Create a dictionary for renaming
rename_dict = {
    'norm_hyper': 'Hyperactivity',
    'norm_ir_agit': 'Irritability/Agitation',
    'norm_leth_soc': 'Lethargy/Social Withdrawal',
    'norm_speech': 'Speech Issues',
    'norm_stereo': 'Stereotypy'
}
```

```python
# heatmap for differences in ABC scores by Disability
abc_by_service_disability = calculate_mean_scores_by_group(demographic_abcs, ['Type of Service', 'Level of Intellectual Disability'])
abc_by_service_disability = abc_by_service_disability[abc_by_service_disability['Level of Intellectual Disability'] != 'None noted']
abc_disability_melted = reshape_data(abc_by_service_disability, ['Type of Service', 'Level of Intellectual Disability'])
abc_disability_pivot = pivot_data(abc_disability_melted, ['Type of Service', 'ABC Score'], 'Level of Intellectual Disability')
abc_disability_diff = calculate_differences(abc_disability_pivot)
abc_disability_diff = rename_and_sort_index(abc_disability_diff, rename_dict)
plot_heatmap(abc_disability_diff, 'Differences in Mean ABC Scores from Intake to Re-Evaluation by Intellectual Disability', 'Level of Intellectual Disability', 'ABC Score', '../output/heatmap_disability.png')
```

```python
# heatmap for differences in ABC scores by gender

abc_by_service_gender = calculate_mean_scores_by_group(demographic_abcs, ['Type of Service', 'Gender'])
abc_melted = reshape_data(abc_by_service_gender, ['Type of Service', 'Gender'])
abc_gender_pivot = pivot_data(abc_melted, ['Type of Service', 'ABC Score'], 'Gender')
abc_gender_diff = calculate_differences(abc_gender_pivot)
abc_gender_diff = rename_and_sort_index(abc_gender_diff, rename_dict)
plot_heatmap(abc_gender_diff, 'Differences in Mean ABC Scores from Intake to Re-Evaluation by Gender', 'Gender', 'ABC Score', '../output/heatmap_gend.png')

```
